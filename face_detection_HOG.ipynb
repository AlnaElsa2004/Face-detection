# Face detection with HOG and Dlib

Face detection using Histogram of Oriented Gradients (HOG) and the dlib library is a widely-used method in computer vision for efficiently detecting faces in images. HOG captures the shape and appearance of objects by computing edge directions and gradients within localized cells, forming a feature descriptor. In dlib, these HOG features are then passed to a Support Vector Machine (SVM) classifier, which determines whether a region contains a face or not. This combination offers an effective balance between speed and accuracy, making it ideal for real-time applications. The method works well under various conditions, such as different lighting and facial expressions, and is less computationally intensive compared to deep learning-based models, while still providing reliable face detection.

Detecting faces with HOG


Loading the image

```python
from google.colab.patches import cv2_imshow
image=cv2.imread('/content/my_new_folder/people2.jpg')
cv2_imshow(image)
```

```python
face_detector_hog=dlib.get_frontal_face_detector()
```

```python
detections=face_detector_hog(image,1)
```

```python
detections,len(detections)
```

There are 14 detections. So the algorithm can detect all 14 faces.

Drawing rectangle boxes around each of the detected faces

```python
for face in detections:
  print(face.left())
  print(face.top())
  print(face.right())
  print(face.bottom()) #The first value indicate the left pointer, second-top pointer,third-right pointer and fourth-bottom pointer
```

```python
for face in detections:
  l,t,r,b=face.left(),face.top(),face.right(),face.bottom()
  cv2.rectangle(image,(l,t),(r,b),(0,255,255),2)
cv2_imshow(image)
```
